```python
# import libraries
import pandas as pd
```

```python
# read all files into their respective segments
seg_1 = pd.read_csv('csv-1700-1830.csv', encoding = "ISO-8859-1", parse_dates=['date(yyyyMMddHHmmss)'], date_format='ISO8601')
seg_2 = pd.read_csv('csv-1831-2000.csv', encoding = "ISO-8859-1", parse_dates=['date(yyyyMMddHHmmss)'], date_format='ISO8601')
seg_3 = pd.read_csv('csv-2001-2131.csv', encoding = "ISO-8859-1", parse_dates=['date(yyyyMMddHHmmss)'], date_format='ISO8601')

# concat vertically as df
df = pd.concat([seg_1, seg_2, seg_3], ignore_index=True)
```

```python
# get shape of individual segments and concatenated dataframe
print('Segment 1 shape:', seg_1.shape)
print('Segment 2 shape:', seg_2.shape)
print('Segment 3 shape:', seg_3.shape)
print('Dataframe shape:', df.shape)
```

```python
# get column names
for col in df.columns:
    print(col, end=' ')
```
